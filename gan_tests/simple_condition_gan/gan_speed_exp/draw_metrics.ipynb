```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import glob
```

```python
legends = []
fig = plt.figure(figsize=(12, 12))
plt.rcParams.update({'font.size': 22})

for single_data in glob.glob('results/*.npy'):
    loaded_metrics = np.load(single_data)
    plt.plot(range(1, len(loaded_metrics)+1), loaded_metrics)
    name = single_data.split('/')[-1].split('.')[0]
    name = " ".join(name.split('-'))
    name = " ".join(name.split('_'))
    legends.append(name)
    print(f'metric={name}, values={loaded_metrics}')
plt.legend(legends)

plt.xlabel(' ')
plt.ylabel('FID')
plt.grid(True)

fig.savefig('mean_train_figure.svg')
plt.show()
```
